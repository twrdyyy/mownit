# MOwNiT Lab4 Symulowane wyżarzanie Filip Twardy

Pseudokod algorytmu:

1.***state*** ← początkowy stan \
2.***cost*** ← cost(***state***) \
3.***i*** ← od ***1*** do ***maksymalna ilosc krokow*** \
4.$\;\;\;\;$***t*** ← wylicz temperature z ***i***-tego kroku \
5.$\;\;\;\;$***new_state*** ← nowy stan wyliczony z obecnego najlepszego \
6.$\;\;\;\;$***new_cost***  ← cost(***new_state***) \
7.$\;\;\;\;$jeżeli zakceptujemy nowy stan: \
8.$\;\;\;\;$$\;\;\;\;$***state*** ← ***new_state*** \
9.$\;\;\;\;$$\;\;\;\;$***cost***  ← ***new_cost***

### Przygotowanie

```python
from generate import uniform_distro, normal_distro, nine_groups
from traveling_salesman import anneal, consectutive_swap, arbitary_swap
from vis import plot_anealing

LOW = 1
HIGH = 1000
N = 1000
MU = 0
SIGMA = 1000
```

### Rozkład normalny

```python
def init_state():
    return normal_distro(MU, SIGMA, N)
```

#### consectutive swap

```python
%%time
states, cost, states, costs = anneal(init_state, consectutive_swap, maxsteps=50000)
plot_anealing(states, costs)
```

#### arbitary swap

```python
%%time
state, cost, states, costs = anneal(init_state, arbitary_swap, maxsteps=50000)
plot_anealing(states, costs)
```

### Rozkład równomierny

```python
def init_state():
    return uniform_distro(LOW, HIGH, N)
```

#### consectutive swap

```python
%%time
state, cost, states, costs = anneal(init_state, consectutive_swap, maxsteps=50000)
plot_anealing(states, costs)
```

#### arbitary swap

```python
%%time
state, cost, states, costs = anneal(init_state, arbitary_swap, maxsteps=50000)
plot_anealing(states, costs)
```

### 9 grup utworzonych z rozkładu równomiernego

```python
def init_state():
    return [y for x in nine_groups(LOW, HIGH, N) for y in x]
```

#### consectutive swap

```python
%%time
state, cost, states, costs = anneal(init_state, consectutive_swap, maxsteps=50000)
plot_anealing(states, costs)
```

#### arbitary swap

```python
%%time
state, cost, states, costs = anneal(init_state, arbitary_swap, maxsteps=50000)
plot_anealing(states, costs)
```

### Wnioski
